```python
import matplotlib.pyplot as plt
import numpy as np

# Pure RSO without magnetism
Co = 1.46
Cr = -0.30
V = -0.46
Ni = 1.87

# # Pure RSO with magnetism
# Co = 1.3893506733797787
# Cr = -5.435543401774682
# V = 0.9289466561113751
# Ni = 1.6749267672077406

# HERSOs without magnetism
run1 = 1.72
run2 = 1.82
run3 = 1.71
run5 = 1.87
run6 = 1.81
run8 = 1.63
run9 = 1.68
run10 = 1.79
run11 = 1.87
run12 = 1.84
run13 = 1.71
run15 = 1.67
run16 = 1.85
run17 = 1.76
run18 = 1.70

# # HERSOs with magnetism
# run1 = 1.4855217725242664
# run2 = 2.4045464158607195
# run3 = 2.0052422777489425
# run5 = 2.9686622710911403
# run6 = 0.06059134974397893
# run8 = 1.7668292827066914
# run9 = 1.4518524496602467
# run10 = 1.626530104646008
# run11 = 1.2967604621328814
# run12 = 1.412088929844752
# run13 = 0.23663088651475256
# run15 = 1.5545475151857195
# run16 = 0.758974098961756
# run17 = 3.019795982773115
# run18 = 2.161172202657613


# runs = np.array([run1, run2, run5, run6, run8, run9, run10, run11, run12, run13, run15, run16, run17, run18])
runs = np.array([run1, run2, run3, run5, run6, run8, run9, run10, run11, run12, run13, run15, run16, run17, run18])

# Create x values and volcano curve
dist = 2.4
x = np.linspace(1.6-dist, 1.6+dist, 1000)
volcano = 1.23 - np.maximum(x, 3.2 - x)

# Initialize the plot with a larger figure size
plt.figure(figsize=(8, 6))

# Plot the volcano curve
plt.plot(x, volcano, color='blue', linewidth=3, zorder=1)

# Plot the data points with labels
plt.scatter(Co, 1.23 - np.maximum(Co, 3.2-Co), s=300, color=[0,1,1], edgecolors='black', linewidth=1.5, zorder=2)
plt.scatter(Cr, 1.23 - np.maximum(Cr, 3.2-Cr), s=300, color=[0,1,0], edgecolors='black', linewidth=1.5, zorder=2)
plt.scatter(Ni, 1.23 - np.maximum(Ni, 3.2-Ni), s=300, color=[0,0,1], edgecolors='black', linewidth=1.5, zorder=2)
plt.scatter(V, 1.23 - np.maximum(V, 3.2-V), s=300, color=[1,1,0], edgecolors='black', linewidth=1.5, zorder=2)
for run in runs:
    plt.scatter(run, 1.23 - np.maximum(run, 3.2-run), marker='^', color='orange', edgecolors='black', s=100, zorder=3)



# Add the LaTeX-formatted x-axis label
plt.xlabel(r'$\Delta G$ (eV)', fontsize=24)
plt.ylabel(r'-$\eta_{OER}$ (V)', fontsize=24)

plt.yticks(fontsize=22)
plt.xticks(fontsize=22)


# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add gridlines along the x-axis
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Optimize layout
plt.tight_layout()
plt.savefig('noMag_full.png', dpi=300, format='png')

# Display the plot
plt.show()

```
